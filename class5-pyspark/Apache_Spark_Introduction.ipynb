```
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Learning_Spark") \
    .getOrCreate()
```

```
data = spark.read.csv('/content/vgsales.csv',inferSchema=True, header=True)
```

```
data.describe(['Publisher', 'Rank', 'Global_Sales']).show()
```
